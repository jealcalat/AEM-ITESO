```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt

# vector de medias
mu = np.array([3.1, 3])

# definir sigma
sigma = np.array([[0.7163, 0.687],
                  [0.687, 0.7163]])
sigma_inv = np.linalg.inv(sigma)
np.random.seed(123)
muestra_bivariada = np.random.multivariate_normal(mu, sigma, 10000)
p1 = np.array([5, 5])
p2 = np.array([5, 2.1])
```

```python
plt.scatter(muestra_bivariada[:, 0], muestra_bivariada[:, 1], alpha=0.2)
plt.scatter(mu[0], mu[1], color='black', label='mu')
plt.scatter(p1[0], p1[1], color='red', label='p1')
plt.scatter(p2[0], p2[1], color='green', label='p2')
plt.legend()
```

```python
dE_p1 = distance.euclidean(mu, p1)
dE_p2 = distance.euclidean(mu, p2)
print(f'DE_p1:{dE_p1}; DE_p2:{dE_p2}')
```

```python
dM_p1 = distance.mahalanobis(p1, mu, sigma_inv)
dM_p2 = distance.mahalanobis(p2, mu, sigma_inv)
print(f'DE_p1:{dM_p1}; DE_p2:{dM_p2}')
```

```python
from scipy.stats import chi2
alfa=0.001
D = 2
valor_crit = chi2.ppf(1 - alfa, D)
valor_crit = np.sqrt(valor_crit)
```

```python
if dM_p1 < valor_crit:
    print(f'p1 con distancia {dM_p1} pertenece a la distribución')
else:
    print(f'p1 con distancia {dM_p1} no pertenece a la distribución')
```

```python
if dM_p2 < valor_crit:
    print(f'p1 con distancia {dM_p2} pertenece a la distribución')
else:
    print(f'p1 con distancia {dM_p2} no pertenece a la distribución')
```
