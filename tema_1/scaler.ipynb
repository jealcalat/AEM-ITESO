```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt # viz
import seaborn as sb # viz
import numpy as np
```

```python
iris = load_iris().data
iris[:,0].min()
```

```python
rango = iris[:,0].max() - iris[:,0].min()
rango
```

```python
media = iris[:,0].mean()
media.round(3)
np.mean(iris[:,0])
```

```python
s = iris[:,0].std()
s
```

```python
z_iris0 = (iris[:,0] - media) / s
z_iris0.mean().round(3)
z_iris0.std()
```

```python
plt.hist(z_iris0, bins=30);
plt.xlim(-5, 10)
plt.hist(iris[:,0], bins=30);
```

Usando `StandardScaler`

```python
scaler = StandardScaler()
z_scaled = scaler.fit(iris).transform(iris)
z_scaled[:,0].mean()
```

```python
plt.hist(z_iris0, bins=30);
plt.xlim(-5, 10)
plt.hist(iris[:,0], bins=30);
plt.hist(z_scaled[:,0], bins=30);
```
