```python
import numpy as np
from scipy.stats import norm
```

```python
mu = 100
sigma = 0.5
lim_inf = 99.5
lim_sup = 100.5

prob_entre = norm.cdf(lim_sup, mu, sigma) - norm.cdf(lim_inf, mu, sigma)
print(prob_entre)
```

```python
lim_inf = 99
lim_sup = 101
prob_inf = norm.cdf(lim_inf, mu, sigma)
prob_sup = 1 - norm.cdf(lim_sup, mu, sigma)
prob_tot = prob_inf + prob_sup
print(prob_tot)
```

```python
(prob_tot)*10000
```

$$
\text{tolerancia} = z\times \sigma_{\text{requerida}}
$$

```python
tol = 0.6
z = norm.ppf(0.975, loc=0, scale=1)
z
```

```python
sigma_requerida = tol / z
print(sigma_requerida)
```

```python
norm.cdf(mu + tol, mu, sigma_requerida) - norm.cdf(mu - tol, mu, sigma_requerida)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
x_values = np.linspace(0, 1, 1000)
y_values = np.linspace(0, 1, 1000)

X, Y = np.meshgrid(x_values, y_values)
f_xy = 1 * X * Y

fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, f_xy, cmap='jet')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('f(x,y')
```
