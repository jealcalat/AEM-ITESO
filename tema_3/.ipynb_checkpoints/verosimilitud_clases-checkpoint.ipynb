## Obtener el MLE en Python usando métodos numéricos

$$
p(k; \theta, n) = \binom{n}{k} \theta^k (1-\theta)^{n-k}
$$

```python
from math import *
from scipy.optimize import minimize
from scipy.optimize import Bounds

# definir binomial

def binom_log_lik(n, k, theta):
    return -(k * np.log(theta) + (n - k) * np.log(1 - theta))

# grid search: definir múltiples valores para el parámetro candidato (theta)
n = 10
k = 9

step = 0.0001

grid = np.linspace(0, 1, int(1 / step))
neg_log_like = binom_log_lik(n, k, grid)
ind_arg_min = np.argmin(neg_log_like)
grid[ind_arg_min]

plt.plot(grid, neg_log_like)
plt.scatter(grid[ind_arg_min], np.min(neg_log_like), c='red')
plt.text(
    grid[ind_arg_min] * 0.8, np.min(neg_log_like) * 3,
    s=f'argmax $l(\\theta)={{{ round(grid[ind_arg_min], 3) }}}$'
)
```

```python
n=1000
k=9
theta_0=0.8

def binom_log_lik(theta, n, k):
    theta = np.clip(theta, 1e-10, 1 - 1e-10)
    # negar para hacerla convexa
    ll = k * np.log(theta) + (n - k) * np.log(1 - theta)
    neg_ll = -ll
    return neg_ll

res = minimize(binom_log_lik, # funcion convexa (minimizar)
               theta_0, # parámetro inicial (en donde comenzar la búsqueda)
               args=(n,k), # constantes; no se buscan, por lo tanto se declaran
               method='L-BFGS-B', # método de búsqueda
               options={'maxiter': 100000}, # si no converge, aumentar la cantidad de iteraciones
               bounds=[(0, 1)]) # restrictiones para hacer la búsqueda de candidatos a theta

print(res.x, k/n)
```

```python
# obtener el MLE usando métodos numéricos de optimización
# 1. definir la fn de log-verosimilitud para la exp
# 2. Usar método de optimización para encontrar el valor en teta* para el cual la función tiene un max
#    Usar el truco de la fn de log-ver negativa para usar minimize

def neg_log_like_expon(lambda_, data):
    like = expon.pdf(data, scale=lambda_)
    log_like = np.log(like)
    return -np.sum(log_like)

lambda0 = 1
result = minimize(neg_log_like_expon, lambda0, args=sim_data, method='L-BFGS-B', bounds=[(0, np.inf)])
result.x
```

## Modelos de mezclas gaussianas


```python
np.random.seed(42)
# simular dos distribuciones gaussianas; obtener una variable aleatoria que provenga de una mezcla de dos guassianas con los
# siguientes parámetros:
n = 10000
mu1 = 150
mu2 = 180

sigma1 = 8
sigma2 = 10

ws = [0.65, 0.35]

s1 = norm.rvs(loc=mu1, scale=sigma1, size=int(n * ws[0]))
s2 = norm.rvs(loc=mu2, scale=sigma2, size=int(n * ws[1]))

smix = np.concatenate([s1, s2])
len(smix)
```

```python
# add kde: kernel density estimation, non-parametric density estimation
from scipy.stats import gaussian_kde

kde = gaussian_kde(smix)
x_grid = np.linspace(120, 240, 1000)
plt.hist(smix, bins=120, density=True, alpha=0.6);
plt.plot(x_grid, kde(x_grid))
plt.axvline(smix.mean(), c='r')
```

```python
help(minimize)
```

```python
def mix_gauss_pdf(x, mu1, mu2, sig1, sig2, w1):
    mix1 = norm.pdf(x, mu1, sig1)
    mix2 = norm.pdf(x, mu2, sig2)
    tot_mix = w1 * mix1 + (1 - w1) * mix2
    return tot_mix

def neg_log_like_mix_gauss(params, x):
    mu1, mu2, sig1, sig2, w1 = params
    # log vero
    log = np.log(mix_gauss_pdf(x, mu1, mu2, sig1, sig2, w1))
    return -np.sum(log)
# inicializar parametros
params0 = [140, 160, 5, 10, 0.85]

result = minimize(neg_log_like_mix_gauss,
                  params0,
                  args=smix,
                  # method='Nelder-Mead'
                  method='TNC',
                  options={'maxfun':1500},
                  bounds=[(0, 350), (0, 350), (0, np.inf), (0, np.inf), (0, 1)]
                 )
result.x.round(3)
# array([158.294, 177.363,   9.315,  12.96 ,   0.336])
```

```python
from scipy.optimize import differential_evolution

def mix_expon_gaussian(params, x):
    l, mu, sig, w = params  # Unpack parameters
    l = min(l, 0.01)
    log_exp = np.log(expon.pdf(x, loc=0, scale=l))
    log_gauss = np.log(norm.pdf(x, mu, sig))
    total_ll = w * log_exp + (1 - w) * log_gauss
    return -np.sum(total_ll)

def wrapper_func(param, *args): 
    return mix_expon_gaussian(param, args[0]) 
bounds = [(0.001, 100), (0, 200), (0, 50), (0.0001, 0.9999)] 

res = differential_evolution(wrapper_func, 
                             bounds,
                             args=(mix_norm_exp,),  # Pass data as tuple in args 
                             maxiter=10000,
                             disp=False)
res.x.round(3)
```

```python
def linear_fun(x, a, b):
    return a + b * x
# params = [a, b, sigma]
def neg_log_like(params, x, y_obs):
    a = params[0]
    b = params[1]
    sig = params[2]

    y_pred = a + b * x
    # epsilon
    error = y_obs - y_pred

    nll = -np.sum(
            np.log( norm.pdf(error, loc=0, scale=sig) + 0.5)
    )
    return nll

```
